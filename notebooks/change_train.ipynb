```python
import os
import pandas as pd
```

```python
df_train = pd.read_csv("../data/collaborative_filtering/train.csv")
df_train
```

```python
def filtrar_por_minimos(df, min_item=3, min_user=3, iterativo=True):
    """
    Filtra un dataframe de interacciones (user, item, rating) eliminando:
    - usuarios con menos de min_user valoraciones
    - ítems con menos de min_item valoraciones

    Parámetros
    ----------
    df : pd.DataFrame
        Debe contener columnas 'user' e 'item' (y opcionalmente 'rating').
    min_item : int
        Mínimo de valoraciones por ítem.
    min_user : int
        Mínimo de valoraciones por usuario.
    iterativo : bool
        Si True, repite el filtrado hasta que no haya más cambios.

    Retorna
    -------
    pd.DataFrame
        DataFrame filtrado.
    """
    out = df.copy()

    if not iterativo:
        usuarios_validos = out["user"].value_counts()
        usuarios_validos = usuarios_validos[usuarios_validos >= min_user].index
        out = out[out["user"].isin(usuarios_validos)]

        items_validos = out["item"].value_counts()
        items_validos = items_validos[items_validos >= min_item].index
        out = out[out["item"].isin(items_validos)]

        return out.reset_index(drop=True)

    # Filtrado iterativo (recomendado)
    while True:
        n_antes = len(out)

        usuarios_validos = out["user"].value_counts()
        usuarios_validos = usuarios_validos[usuarios_validos >= min_user].index
        out = out[out["user"].isin(usuarios_validos)]

        items_validos = out["item"].value_counts()
        items_validos = items_validos[items_validos >= min_item].index
        out = out[out["item"].isin(items_validos)]

        if len(out) == n_antes:
            break

    return out.reset_index(drop=True)

# Ejemplo de uso:
df_train_filtrado = filtrar_por_minimos(df_train, min_item=0, min_user=2)
df_train_filtrado
```

```python
df_train_filtrado.to_csv("../data/cut_1_0/train.csv", index=False)
```
